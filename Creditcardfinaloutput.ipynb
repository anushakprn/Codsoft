```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
import zipfile

# Specify the path to your zip file
zip_file_path = r'C:\Users\anush\Desktop\final.zip'

# Extract the contents of the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Assuming the zip file contains only one CSV file, you can use the first one
    csv_file_name = zip_ref.namelist()[0]
    zip_ref.extract(csv_file_name, 'temp_directory')  # Extract to a temporary directory

# Read the CSV file into a Pandas DataFrame
credit_card_data = pd.read_csv(f'temp_directory/{csv_file_name}')

# Display the DataFrame
print(credit_card_data)

# Optionally, you can remove the temporary directory
# import shutil
# shutil.rmtree('temp_directory')
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python

new_dataset.head()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python

print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python

model.fit(X_train, Y_train)
```

```python

# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
     
```

```python


print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
